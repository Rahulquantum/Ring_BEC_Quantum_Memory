#### Configure matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import MaxNLocator
import numpy as np

# enable TeX fonts
plt.rcParams['text.usetex'] = True

# Set the global font size for all legends
mpl.rcParams['legend.fontsize'] = 14

# Set the global font size for all axis labels
mpl.rcParams['axes.labelsize'] = 12

# Set the global font style for all axis labels
plt.rcParams['font.family'] = 'Times New Roman'
```

#### Cavity decay rate, $\gamma_0$

```python
kappa0 = 2*np.pi*1000
```

#### Collect data

```python
# Full path to the file in the current directory
filename1 = './fid_vs_Lp.txt'
filename11 = './fid_vs_Lp_with_int.txt'
filename2 = './fid_vs_l.txt'
filename22 = './fid_vs_l_with_int.txt'
filename3 = './fid_vs_pulse_width.txt'

# Load data assuming two columns (x and y), handling spaces, tabs, or commas
data1 = np.loadtxt(filename1, delimiter=None)
data11 = np.loadtxt(filename11, delimiter=None)
data2 = np.loadtxt(filename2, delimiter=None)
data22 = np.loadtxt(filename22, delimiter=None)
data3 = np.loadtxt(filename3, delimiter=None)

# Split into x and y components
x1 = data1[:,0]
x11 = data11[:,0]
x2 = data2[:,0]
x22 = data22[:,0]
x3 = data3[:,0]
y1 = data1[:, 1]
y11 = data11[:, 1]
y2 = data2[:, 1]
y22 = data22[:, 1]
y3 = data3[:, 1]
```

#### Plot 1

```python
# Plot the data
plt.figure(figsize=(7, 5))
plt.plot(x1, y1/100, marker='o', linestyle='-', color='b', label = r'$\tilde{g}/\gamma_0 = 0$')#, label = r'$\tilde{G}/\gamma_0 = 4$')
plt.plot(x11, y11/100, marker='s', linestyle='-', color='r', label = r'$\tilde{g}/\gamma_0 \sim 1.34\times10^{-6}$')#, label = r'$\tilde{G}/\gamma_0 = 4$')
plt.xlabel(r'$L_p$', fontsize = 24)
plt.ylim(0.75,0.95)

plt.ylabel(r'$\mathcal{F}$', fontsize = 24)
plt.tick_params(labelsize=20)
plt.grid(False)

# legends 
plt.legend(fontsize = 23, frameon=False,loc='upper right') 

ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.xaxis.offsetText.set_fontsize(20)

plt.tight_layout()

filename = "figure3i.svg"
plt.savefig(filename, format='svg', transparent=True)

plt.show()
```

#### Plot 2

```python
# Plot the data
plt.figure(figsize=(7, 5))
plt.plot(x2, y2/100, marker='o', linestyle='-', color='b', label = r'$\tilde{g}/\gamma_0 = 0$')
plt.plot(x22, y22/100, marker='s', linestyle='-', color='r', label = r'$\tilde{g}/\gamma_0 \sim 1.34\times10^{-6}$')
plt.xlabel(r'$\ell$', fontsize = 24)
plt.xticks(x2)
plt.ylim(0.75,0.95)

plt.ylabel(r'$\mathcal{F}$', fontsize = 24)
plt.tick_params(labelsize=20)
plt.grid(False)

# legends
plt.legend(fontsize = 23, frameon=False,loc='upper right') 

ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)

ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.xaxis.offsetText.set_fontsize(20)

plt.tight_layout()

filename = "figure3ii.svg"
plt.savefig(filename, format='svg', transparent=True)


plt.show()
```
