#### Libraries

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import rc
from matplotlib import rcParams
import numpy as np

rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['mathtext.fontset'] = 'stix'
```

### Cavity decay rate, $\gamma_0$

```python
kappa0 = 2*np.pi*1000
```

#### Plot

```python
# Full path to the file in the current directory
filename0 = './G_equal_4_log_neg.txt'
filename1 = './G_equal_8_log_neg.txt'

# Load data assuming two columns (x and y), handling spaces, tabs, or commas
data0 = np.loadtxt(filename0, delimiter=None)
data1 = np.loadtxt(filename1, delimiter=None)

# x data
x0 = data0[:, 0]

# y data
y0 = data0[:, 1]
y1 = data1[:, 1]

# Create one main plot
fig, ax = plt.subplots(figsize=(7, 5))

# Plot primary axis lines
ax.plot(x0/kappa0, y0, linestyle='-', color='b', label = r'$\tilde{G}/\gamma_0 = 4$', linewidth = 4)
ax.plot(x0/kappa0, y1, linestyle=':', color='g', label = r'$\tilde{G}/\gamma_0 = 8$', linewidth = 4)

# Legends
ax.legend(fontsize = 24, loc='upper right',  fancybox=False, framealpha=0)#, bbox_to_anchor=(1, 1))#, fancybox=True, framealpha=1)   #

# Set labels and title
ax.set_ylabel(r'$\mathcal{E}_N$', fontsize = 28)
ax.set_xlabel('Storage time (s)', fontsize = 28)

# axis ticks
xticks = [0, 0.2, 0.4, 0.6]
yticks = [0.2, 0.4, 0.6, 0.8]
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.tick_params(axis='both', labelsize=24)

# save the plot to a file
filename = "figure3iiia.svg"
plt.savefig(filename, format='svg', transparent=True)

plt.show()
```
