#### Libraries

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np

rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['mathtext.fontset'] = 'stix'
```

#### Cavity decay rate, $\gamma_0$

```python
kappa0 = 2*np.pi*1000
```

#### Plot

```python
# Full path to the file in the current directory
filename1 = './G_equal_4_ent.txt'
filename2 = './G_equal_8_ent.txt'

# Load data assuming two columns (x and y), handling spaces, tabs, or commas
data1 = np.loadtxt(filename1, delimiter=None)
data2 = np.loadtxt(filename2, delimiter=None)

# x data
x = data1[:, 0]
# y data
y1 = data1[:, 1]
y2 = data2[:, 1]

# Create one main plot
fig, ax = plt.subplots(figsize=(7, 5))

# Plot primary axis lines
ax.plot(x/kappa0, y1, linestyle='-', color='b', label = r'$\tilde{G}/\gamma_0 = 4$', linewidth = 4)
ax.plot(x/kappa0, y2, linestyle=':', color='g', label = r'$\tilde{G}/\gamma_0 = 8$', linewidth = 4)
ax.axhline(y = 2/5, linestyle='--', color='r', linewidth = 4)  # Classical limit


# Legends
ax.legend(fontsize = 24, loc='upper right', fancybox=False, framealpha=0, bbox_to_anchor=(1, 1.05))


# Set labels and title
ax.set_xlabel('Storage time (s)', fontsize = 28)
ax.set_ylabel(r'$\mathcal{F}$', fontsize = 28)

# axes ticks
xticks = [0, 0.2,0.4, 0.6]
yticks = [0.5, 0.6,0.7, 0.8, 0.9]
ax.tick_params(axis='both', labelsize=24)

ax.set_ylim(2/5 - 0.1,1)

#------------------------------------
#       Texts
#------------------------------------
ax.text(0.2, 2/5 + 0.05, 'Classical limit', fontsize=28, color='Magenta')

plt.tight_layout()

# save the plot to a file
filename = "figure3iiib.svg"
plt.savefig(filename, format='svg', transparent=True)


plt.show()
```
