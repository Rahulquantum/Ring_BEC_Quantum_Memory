#### Libraries

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import rcParams
import numpy as np

rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']  # Or 'DejaVu Serif', 'Georgia'
rcParams['mathtext.fontset'] = 'stix'
```

#### Cavity decay rate, $\gamma_0$

```python
kappa_0 = 2*np.pi*1000
```

#### Plots

```python
import matplotlib.pyplot as plt

# Full path to the file in the current directory
filename1 = './G_equal_4.txt'
filename2 = './G_equal_8.txt'

# Load data assuming two columns (x and y), handling spaces, tabs, or commas
data1 = np.loadtxt(filename1, delimiter=None)
data2 = np.loadtxt(filename2, delimiter=None)

# x and y data
x = data1[:, 0]
y1 = data1[:, 1]
y2 = data2[:, 1]

# Create one main plot
fig, ax = plt.subplots(figsize=(8, 5))

# Plot primary axis lines
ax.plot(x/kappa_0, y1, linestyle='-', color='b', label = r'$\tilde{G}/\gamma_0 = 4$', linewidth = 4)
ax.plot(x/kappa_0, y2, linestyle=':', color='g', label = r'$\tilde{G}/\gamma_0 = 8$', linewidth = 4)

# horizontal limit
ax.axhline(y = 2/3, linestyle='--', color='r', linewidth = 3)

# legends
legend = ax.legend(fontsize = 28, loc='upper right', bbox_to_anchor=(1, 1.05), fancybox=True, framealpha=0)


# Set labels and title
ax.set_xlabel('Storage time (s)', fontsize = 30)
ax.set_ylabel(r'$\mathcal{F}$', fontsize = 30)

# axes ticks
ax.tick_params(axis='both', labelsize=27)

# axes limits
ax.set_ylim(0.60,0.98)

#------------------------------------
#       Texts
#------------------------------------
ax.text(0.5, 2/3 + 0.02, 'Classical limit', fontsize=28, color='Magenta')

plt.tight_layout()

# save the plot to a file
filename = "figure2iii.svg"
plt.savefig(filename, format='svg', transparent=True)


plt.show()
```
